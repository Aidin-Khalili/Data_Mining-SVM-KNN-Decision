KNN on data sets (Iris, Ecoli, SPECTF heart)
Links:
1. SPECTF heart: https://archive.ics.uci.edu/ml/machine-learning-databases/spect/SPECTF.train
2. Iris: https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data
3. Ecoli: https://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

def delete_DublicateInstances(theDataset):
    theDataset.drop_duplicates(inplace=True)

def handle_MissingValues(theDataSet):
    for i in range(len(theDataSet.columns)):
        theDataSet.iloc[:,i].fillna(theDataSet.iloc[:,i].mean(), inplace=True)
  
```

```python
#Iris dataSet 
url_iris = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
irisDataSet = pd.read_csv(url_iris, names=['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'class'])
delete_DublicateInstances(irisDataSet)
scaler_irisDataSet = StandardScaler()
scaler_irisDataSet.fit(irisDataSet.drop('class', axis=1))
scaled_irisDataSet = scaler_irisDataSet.transform(irisDataSet.drop('class', axis=1))
x_irisDataSet = scaled_irisDataSet
y_irisDataSet = irisDataSet['class']
x_train_irisDataSet, x_test_irisDataSet, y_train_irisDataSet, y_test_irisDataSet = train_test_split(x_irisDataSet, y_irisDataSet, test_size=0.3)
knn_irisDataSet = KNeighborsClassifier(n_neighbors=1)
knn_irisDataSet.fit(x_train_irisDataSet, y_train_irisDataSet)
errorRate_irisDataSet =[]
for i in range(1,40):
    knn_irisDataSet = KNeighborsClassifier(n_neighbors=i)
    knn_irisDataSet.fit(x_train_irisDataSet, y_train_irisDataSet)
    prediction_irisDataSet = knn_irisDataSet.predict(x_test_irisDataSet)
    errorRate_irisDataSet.append(np.mean(prediction_irisDataSet != y_test_irisDataSet))
fig = plt.figure(figsize=(25,20))
plt.plot(range(1,40), errorRate_irisDataSet, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
```

```python
#BezdekIris dataSet 
url_bezdekIris = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/bezdekIris.data'
bezdekIrisDataSet = pd.read_csv(url_bezdekIris, names=['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'class'])
delete_DublicateInstances(bezdekIrisDataSet)
scaler_bezdekIrisDataSet = StandardScaler()
scaler_bezdekIrisDataSet.fit(bezdekIrisDataSet.drop('class', axis=1))
scaled_bezdekIrisDataSet = scaler_bezdekIrisDataSet.transform(bezdekIrisDataSet.drop('class', axis=1))
x_bezdekIrisDataSet = scaled_bezdekIrisDataSet
y_bezdekIrisDataSet = bezdekIrisDataSet['class']
x_train_bezdekIrisDataSet, x_test_bezdekIrisDataSet, y_train_bezdekIrisDataSet, y_test_bezdekIrisDataSet = train_test_split(x_bezdekIrisDataSet, y_bezdekIrisDataSet, test_size=0.3)
knn_bezdekIrisDataSet = KNeighborsClassifier(n_neighbors=1)
knn_bezdekIrisDataSet.fit(x_train_bezdekIrisDataSet, y_train_bezdekIrisDataSet)
errorRate_bezdekIrisDataSet =[]
for i in range(1,40):
    knn_bezdekIrisDataSet = KNeighborsClassifier(n_neighbors=i)
    knn_bezdekIrisDataSet.fit(x_train_bezdekIrisDataSet, y_train_bezdekIrisDataSet)
    prediction_bezdekIrisDataSet = knn_bezdekIrisDataSet.predict(x_test_bezdekIrisDataSet)
    errorRate_bezdekIrisDataSet.append(np.mean(prediction_bezdekIrisDataSet != y_test_bezdekIrisDataSet))
fig = plt.figure(figsize=(25,20))
plt.plot(range(1,40), errorRate_bezdekIrisDataSet, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
```

```python
#Ecoli dataSet 
url_ecoli = 'https://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data'
ecoliDataSet = pd.read_csv(url_ecoli, names=['Sequence Name', 'mcg', 'gvh', 'lip', 'chg', 'aac', 'alm1', 'alm2', 'class'], sep='\s+')
delete_DublicateInstances(ecoliDataSet)
scaler_ecoliDataSet = StandardScaler()
scaler_ecoliDataSet.fit(ecoliDataSet.drop(columns=['Sequence Name','class'], axis=1))
scaled_ecoliDataSet = scaler_ecoliDataSet.transform(ecoliDataSet.drop(columns=['Sequence Name','class'], axis=1))
x_ecoliDataSet = scaled_ecoliDataSet
y_ecoliDataSet = ecoliDataSet['class']
x_train_ecoliDataSet, x_test_ecoliDataSet, y_train_ecoliDataSet, y_test_ecoliDataSet = train_test_split(x_ecoliDataSet, y_ecoliDataSet, test_size=0.3)
knn_ecoliDataSet = KNeighborsClassifier(n_neighbors=1)
knn_ecoliDataSet.fit(x_train_ecoliDataSet, y_train_ecoliDataSet)
errorRate_ecoliDataSet =[]
for i in range(1,40):
    knn_ecoliDataSet = KNeighborsClassifier(n_neighbors=1)
    knn_ecoliDataSet.fit(x_train_ecoliDataSet, y_train_ecoliDataSet)
    prediction_ecoliDataSet = knn_ecoliDataSet.predict(x_test_ecoliDataSet)
    errorRate_ecoliDataSet.append(np.mean(prediction_ecoliDataSet != y_test_ecoliDataSet))
fig = plt.figure(figsize=(25,20))
plt.plot(range(1,40), errorRate_ecoliDataSet, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
```

```python
#SPECTF Heart DataSet 
url_train_SPECTFDataSet = 'https://archive.ics.uci.edu/ml/machine-learning-databases/spect/SPECTF.train'
train_SPECTFDataSet = pd.read_csv(url_train_SPECTFDataSet, names=['OVERALL_DIAGNOSIS', 'F1R', 'F1S', 'F2R', 'F2S', 'F3R', 'F3S', 'F4R', 'F4S', 'F5R', 'F5S', 'F6R', 'F6S', 'F7R', 'F7S', 'F8R', 'F8S', 'F9R', 'F9S', 'F10R', 'F10S', 'F11R', 'F11S', 'F12R', 'F12S', 'F13R', 'F13S', 'F14R', 'F14S', 'F15R', 'F15S', 'F16R', 'F16S', 'F17R', 'F17S', 'F18R', 'F18S', 'F19R', 'F19S', 'F20R', 'F20S', 'F21R', 'F21S', 'F22R', 'F22S'])
url_test_SPECTFDataSet = 'https://archive.ics.uci.edu/ml/machine-learning-databases/spect/SPECTF.test'
test_SPECTFDataSet = pd.read_csv(url_test_SPECTFDataSet, names=['OVERALL_DIAGNOSIS', 'F1R', 'F1S', 'F2R', 'F2S', 'F3R', 'F3S', 'F4R', 'F4S', 'F5R', 'F5S', 'F6R', 'F6S', 'F7R', 'F7S', 'F8R', 'F8S', 'F9R', 'F9S', 'F10R', 'F10S', 'F11R', 'F11S', 'F12R', 'F12S', 'F13R', 'F13S', 'F14R', 'F14S', 'F15R', 'F15S', 'F16R', 'F16S', 'F17R', 'F17S', 'F18R', 'F18S', 'F19R', 'F19S', 'F20R', 'F20S', 'F21R', 'F21S', 'F22R', 'F22S'])
delete_DublicateInstances(train_SPECTFDataSet)
delete_DublicateInstances(test_SPECTFDataSet)
x_train_SPECTFDataSet = train_SPECTFDataSet.drop(columns=['OVERALL_DIAGNOSIS'], axis=1)
y_train_SPECTFDataSet = train_SPECTFDataSet['OVERALL_DIAGNOSIS']
x_test_SPECTFDataSet = test_SPECTFDataSet.drop(columns=['OVERALL_DIAGNOSIS'], axis=1)
y_test_SPECTFDataSet = test_SPECTFDataSet['OVERALL_DIAGNOSIS']
handle_MissingValues(x_train_SPECTFDataSet)
handle_MissingValues(x_test_SPECTFDataSet)
scaler_train_SPECTFDataSet = StandardScaler()
scaler_train_SPECTFDataSet.fit(x_train_SPECTFDataSet)
scaled_train_SPECTFDataSet = scaler_train_SPECTFDataSet.transform(x_train_SPECTFDataSet)
scaler_test_SPECTFDataSet = StandardScaler()
scaler_test_SPECTFDataSet.fit(x_test_SPECTFDataSet)
scaled_test_SPECTFDataSet = scaler_test_SPECTFDataSet.transform(x_test_SPECTFDataSet)
knn_SPECTFDataSet = KNeighborsClassifier(n_neighbors=1)
knn_SPECTFDataSet.fit(scaled_train_SPECTFDataSet, y_train_SPECTFDataSet)
errorRate_SPECTFDataSet =[]
for i in range(1,40):
    knn_SPECTFDataSet = KNeighborsClassifier(n_neighbors=1)
    knn_SPECTFDataSet.fit(scaled_train_SPECTFDataSet, y_train_SPECTFDataSet)
    prediction_SPECTFDataSet = knn_SPECTFDataSet.predict(scaled_test_SPECTFDataSet)
    errorRate_SPECTFDataSet.append(np.mean(prediction_SPECTFDataSet != y_test_SPECTFDataSet))
fig = plt.figure(figsize=(25,20))
plt.plot(range(1,40), errorRate_SPECTFDataSet, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
```
